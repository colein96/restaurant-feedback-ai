This project explore how AI can analyze customer reviews to identify trends and improve restaurant service.

The goal is extract:
-positive feedback
-negative feedback
-improvement suggestions

```python
reviews = [
"Great food and friendly staff",
"Amazing pasta but the service was slow",
"Very noisy environment but good cocktails",
"Excellent pizza and quick service",
"Food was cold and the waiter forgot our order",
"Nice atmosphere but very expensive",
"Fantastic experience and delicious desserts",
"Long waiting time and poor service",
"Very fresh ingredients and great menu",
"Staff was rude and the table was dirty"
]

print("Number of reviews:", len(reviews))
```

```python
positive_words = ["great","amazing","excellent","fantastic","fresh","delicious"]
negative_words = ["slow","cold","rude","dirty","poor","expensive","noisy"]

positive = []
negative = []

for review in reviews:
    if any(word in review.lower() for word in positive_words):
        positive.append(review)
    if any(word in review.lower() for word in negative_words):
        negative.append(review)

print("Positive feedback:")
for r in positive:
    print("-", r)

print("\nNegative feedback:")
for r in negative:
    print("-", r)
```

This script analyzes restaurant reviews and separates
positive and negative feedback using basic NLP logic.


```python
service_keywords = ["slow", "rude", "waiting", "service"]
food_keywords = ["cold", "pizza", "pasta", "food"]
environment_keywords = ["noisy", "dirty", "atmosphere"]

service_issues = []
food_feedback = []
environment_feedback = []

for review in reviews:
    r = review.lower()

    if any(word in r for word in service_keywords):
        service_issues.append(review)

    if any(word in r for word in food_keywords):
        food_feedback.append(review)

    if any(word in r for word in environment_keywords):
        environment_feedback.append(review)

        print("Service related feedback:")
for r in service_issues:
    print("-", r)

print("\nFood related feedback:")
for r in food_feedback:
    print("-", r)

print("\nEnvironment related feedback:")
for r in environment_feedback:
    print("-", r)
```

```python
print("\nSuggested Improvements:")

if len(service_issues) > 1:
    print("- Improve service speed and staff training")

if len(environment_feedback) > 1:
    print("- Improve restaurant cleanliness or noise control")

if len(food_feedback) > 1:
    print("- Review food preparation consistency")
```

This tool simulates how AI can help restaurant managers
analyze customer feedback and identify improvement areas.

```python
import matplotlib.pyplot as plt

categories = ["Service Issues", "Food Feedback", "Environment Issues"]
counts = [len(service_issues), len(food_feedback), len(environment_feedback)]

plt.bar(categories, counts)
plt.title("Restaurant Feedback Analysis")
plt.xlabel("Feedback Categories")
plt.ylabel("Number of Mentions")

plt.show()
```

This visualization helps restaurant managers quickly identify
the most common customer feedback trends.

```python
from textblob import TextBlob

for review in reviews:
    blob = TextBlob(review)
    if blob.sentiment.polarity > 0:
        sentiment = "Positive"
    elif blob.sentiment.polarity < 0:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    print(f"{review} -> {sentiment}")
```
